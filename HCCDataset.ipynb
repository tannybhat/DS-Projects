```python
def generate_model(d,s):
    
        data = d
        true_labels = data['Class']
        data_r = data.drop(["Class"], 1)
        from sklearn.feature_selection import RFE
        from sklearn.linear_model import LinearRegression
        import xgboost as xgb

        #Selecting Features using RFE
        model = LinearRegression()
        rfe = RFE(model, 4)
        y=rfe.fit(data_r,true_labels)
        x = rfe.fit_transform(data_r, true_labels)
        data_rfe = pd.DataFrame(x)
        data_rfe.head()


    #Converting Selected FEatures to df
        Selected_features = y.support_
        dropped_names = []
        for i in range(len(Selected_features)):
            if Selected_features[i] == False:
                dropped_names.append(data_r.columns[i])

                #Adding in Features Selected Manually with plots and correlation
        dropped_names.remove('Gender') 
        dropped_names.remove('Alcohol')
        dropped_names.remove('Smoking')
        dropped_names.remove('Varices')
        dropped_names.append('HBeAg')
        #dropped_names.append('null_values')
        data_new =  data_r.drop(dropped_names,axis=1)
        data_new.head()
        data_x = data_new
        from sklearn.preprocessing import MinMaxScaler, StandardScaler
        min_max_scaler = MinMaxScaler()
        data_new = pd.DataFrame(min_max_scaler.fit_transform(data_new))

        data_new.columns = data_x.columns
        data_new.head()
        train_X =data_new[:140]
        train_Y =true_labels[:140]

        test_X = data_new[140:]
        test_Y = true_labels[140:]
        from sklearn import svm
        clf = svm.SVC(kernel='linear', C = 1)
        clf.fit(train_X,train_Y)
        y_pred1 = clf.predict(test_X)

        from sklearn import metrics

        print(f'SVC Predictions for {s}:')
        # Model Accuracy, how often is the classifier correct?
        print("Accuracy:",metrics.accuracy_score(test_Y, y_pred1)*100,"%")
        print("F1 Score:",metrics.f1_score(test_Y, y_pred1)*100,"%")
        print("Precision:",metrics.precision_score(test_Y, y_pred1)*100,"%")
        print("Recall:",metrics.recall_score(test_Y, y_pred1)*100,"%")
        print("")

        import xgboost 
        xgb = xgboost.XGBClassifier(n_estimators=100, learning_rate=0.01, gamma=0.1, subsample=0.75,
                                   colsample_bytree=1, max_depth=10)
        xgb.fit(train_X,train_Y)
        predictions = xgb.predict(test_X)
        print(f'XGboost Predictions for {s}:')
        print("Accuracy:",metrics.accuracy_score(test_Y, predictions)*100,"%")
        print("F1 Score:",metrics.f1_score(test_Y, predictions)*100,"%")
        print("Precision:",metrics.precision_score(test_Y, predictions)*100,"%")
        print("Recall:",metrics.recall_score(test_Y, predictions)*100,"%")
        return dropped_names,data_new
dropped_names,data_new = generate_model(test_maxna,"Max Replacement")
```

b.	Can you identify if you model is underfitting or overfitting? (Hint use cross validation error and in-sample error plot to identify high bias and high variance.) Plot the relationships.

```python
degree = [i for i in range(10)]
accuracy = []

for i in degree: 
    X_transform = pd.DataFrame(PolynomialFeatures(i).fit_transform(data_new))
    train_X =X_transform[:140]
    train_Y =true_labels[:140]

    test_X = X_transform[140:]
    test_Y = true_labels[140:]

    clf = svm.SVC(kernel='linear', C = 1)
    clf.fit(train_X,train_Y)
    y_pred1 = clf.predict(test_X)
    accuracy.append(1 - metrics.accuracy_score(test_Y, y_pred1))

plt.plot(degree,accuracy,linestyle='--', marker='o', color='red')
plt.title('Degree Vs Accuracy')
plt.xlabel('degree')
plt.ylabel('Error')
plt.show()
```

```python
from sklearn.model_selection import validation_curve



param_range = np.logspace(-5, 8, 5)
train_scores, test_scores = validation_curve(
    svm.SVC(), X_transform, true_labels, param_name="gamma", param_range=param_range,
    cv=5, scoring="accuracy", n_jobs=1)
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

plt.title("Validation Curve with SVM")
plt.xlabel(r"$\gamma$")
plt.ylabel("Score")
plt.ylim(0.0, 1.1)
lw = 2
plt.semilogx(param_range, train_scores_mean, label="Training score",
             color="darkorange", lw=lw)
plt.fill_between(param_range, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.2,
                 color="darkorange", lw=lw)
plt.semilogx(param_range, test_scores_mean, label="Cross-validation score",
             color="navy", lw=lw)
plt.fill_between(param_range, test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std, alpha=0.2,
                 color="navy", lw=lw)
plt.legend(loc="best")
plt.show()
```

For very low values of gamma, you can see that both the training score and the validation score are low. This is called underfitting. 
Medium values of gamma will result in high values for both scores, i.e. the classifier is performing fairly well. If gamma is too high, the classifier will overfit, which means that the training score is good but the validation score is poor.
